PREDICTING THE OPTIMUM VALUE OF K FOR CLUSTERING

```python
x = df.iloc[:, [0, 1, 2, 3]].values
```

```python
plt.scatter(x[:,0],x[:,1])
```

```python
k=KMeans(n_clusters=2)
y=k.fit_predict(x)
```

```python
plt.scatter(x[:,0],x[:,1],c=y,cmap='rainbow')
```

```python
inertia=[]
for i in range(1,11):
    k=KMeans(n_clusters=i)
    k.fit(x)
    inertia.append(k.inertia_)
plt.plot(range(1,11),inertia)
plt.title('Optimum No of Clusters')
plt.xlabel('No of Clusters')
plt.ylabel('Inertia')
```

```python
k=KMeans(n_clusters=3)
y=k.fit_predict(x)
plt.scatter(x[:,0],x[:,1],c=y,cmap='rainbow')
```
